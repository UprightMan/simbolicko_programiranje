```python
import random
random.seed
```

```python
n = 18300
m = 11
skill = random.random() * 100
luck = random.random() * 100
score = 0.95 * skill + luck * 0.05
```

```python
skill, luck, score
```

```python
successful_applicants_all_runs = []
successful_applicants_all_runs_skill = []
for j in range(1000):
    applicants_score = []
    applicants_luck = []
    applicants_skill = []
    for i in range(m):
        skill = random.random() * 100
        luck = random.random() * 100
        score = 0.95 * skill + luck * 0.05
        applicants_score.append(score)
        applicants_luck.append(luck)
        applicants_skill.append(skill)
    for i in range(n - m):
        skill = random.random() * 100
        luck = random.random() * 100
        score = 0.95 * skill + luck * 0.05
        if min(applicants_score) < score:
            applicants_luck[applicants_score.index(min(applicants_score))] = luck
            applicants_skill[applicants_score.index(min(applicants_score))] = skill
            applicants_score[applicants_score.index(min(applicants_score))] = score
    successful_applicants_all_runs.extend(applicants_luck)
    successful_applicants_all_runs_skill.extend(applicants_skill)
```

```python
def Average(lst):
    return sum(lst) / len(lst)
a = Average(successful_applicants_all_runs)
a
```

```python
from sympy import *
```

```python
def f(x, y):
    x += 1
    y += 1
    if x >= y:
        return root(log(y, Symbol('p')), Symbol('p'))
    elif y > x:
        return root(log(x, Symbol('p')), Symbol('p'))
```

```python
def fun(n):
    M = Matrix(n, n, f)
    M = M ** n
    simplify(M)
    return M
```

```python
fun(3)
```

```python
def fun(n):
    tr = 0
    p = Symbol('p', real=True, positive = True)
    for i in range(1, n+1):
        tr += root(log(i, p), p)
    return tr
```

```python
minimum(fun(5), Symbol('p', real = True, positive = True))
```
