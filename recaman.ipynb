```python
def recaman(n: int):
    sequence = []
    for i in range(n):
        if i == 0:
            sequence.append(0)
        elif sequence[i-1] - i < 0  or  sequence[i-1] - i in sequence:
            sequence.append(sequence[i-1] + i)
        else:
            sequence.append(sequence[i-1] - i)
    return sequence
```

```python
import matplotlib
import matplotlib.pyplot
from matplotlib import patches
from matplotlib.axes import Axes
```

```python
def circles(axes: Axes, sequence: list):
    for i, j in enumerate(sequence[1:], 1):
        if i % 2 == 0:
            angles = (0, 180)
        else:
            angles = (180, 0)
        arc = patches.Arc(((float(j) + float(sequence[i-1]))/2, 0), abs(j - sequence[i - 1]), abs(j - sequence[i - 1]), theta1=angles[0], theta2=angles[1], linewidth = 0.5)
        axes.add_patch(arc)
```

```python
sequence = recaman(100)
fig, ax = matplotlib.pyplot.subplots()
max_diameter = max(sequence)
ax.set_frame_on(False)
ax.get_yaxis().set_visible(False)
ax.get_xaxis().set_visible(False)
ax.set_xlim(0 - 1, max(sequence) + 1)
ax.set_ylim(-(max_diameter / 2 + 1), max_diameter / 2 + 1)
ax.set_aspect('equal', adjustable='box')
circles(ax, sequence)
matplotlib.pyplot.savefig('Recaman.pdf', dpi = 500, facecolor = 'w')
```
